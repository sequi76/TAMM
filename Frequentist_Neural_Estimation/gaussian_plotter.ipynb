```python
import numpy as np
import torch
import matplotlib.pyplot as plt
this_dir = "./gaussian_scripts/"
msd_signal_means = torch.tensor(torch.load(this_dir + 'signal_means_final.pt'))
msd_signal_covs = torch.tensor(torch.load(this_dir + 'signal_covs_final.pt'))
msd_background_means = torch.tensor(torch.load(this_dir + 'background_means_final.pt'))
msd_background_covs = torch.tensor(torch.load(this_dir + 'background_covs_final.pt'))

Nb0 = 100000
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
CUT = 2

def matrix_from_params(means, covs):
    msd_matrix = []
    for i in range(len(means)):
        mean = means[i]
        cov = covs[i]
        msd_matrix.append(torch.tensor(np.random.multivariate_normal(mean.numpy(), cov.numpy(), Nb0)).float())
    return msd_matrix

signal_msds = matrix_from_params(msd_signal_means, msd_signal_covs)
background_msds = matrix_from_params(msd_background_means, msd_background_covs)

td_signal_mean = [0, 0]
td_signal_cov = [[1, 0], [0, 1]]
td_background_mean = [0, 1]
td_background_cov = [[3, 0.4], [0.4, 3]]

obs_signal_raw = torch.tensor(np.random.multivariate_normal(td_signal_mean, td_signal_cov, 100000)).float()
obs_signal_raw = obs_signal_raw[(obs_signal_raw[:, 0] >= -CUT) & (obs_signal_raw[:, 0] <= CUT) & (obs_signal_raw[:, 1] >= -CUT) & (obs_signal_raw[:, 1] <= CUT)]
obs_signal_p = obs_signal_raw

obs_background_raw = torch.tensor(np.random.multivariate_normal(td_background_mean, td_background_cov, 100000)).float()
obs_background_raw = obs_background_raw[(obs_background_raw[:, 0] >= -CUT) & (obs_background_raw[:, 0] <= CUT) & (obs_background_raw[:, 1] >= -CUT) & (obs_background_raw[:, 1] <= CUT)]
obs_background_p = obs_background_raw
```

```python
def normalize_prob(x,epsilon=0.0001):
    x = x + epsilon
    x = np.asarray(x, dtype=np.float64)
    s = x.sum(dtype=np.float64)
    y = x / s
    i = y.argmax()                     # adjust the largest to keep all positive
    y[i] += 1.0 - y.sum(dtype=np.float64)
    return y


def flat_histogramdd(arr, bins):
    edges = [np.asarray(b) for b in bins]
    return np.histogramdd(arr, bins=edges)[0].ravel()

def hellinger_distance(p, q):
    return np.sqrt(0.5 * np.sum((np.sqrt(p) - np.sqrt(q)) ** 2))

mybinning = [np.linspace(-2, 2, 11), np.linspace(-2, 2, 11)]
obs_signal_hist = flat_histogramdd(obs_signal_p.numpy(), bins=mybinning)
obs_signal_hist = normalize_prob(obs_signal_hist)
obs_background_hist = flat_histogramdd(obs_background_p.numpy(), bins=mybinning)
obs_background_hist = normalize_prob(obs_background_hist)

signal_hds = []
background_hds = []

for signal_msd in signal_msds:
    signal_cut_mask = (signal_msd[:, 0] >= -CUT) & (signal_msd[:, 0] <= CUT) & (signal_msd[:, 1] >= -CUT) & (signal_msd[:, 1] <= CUT)
    signal_msd_cut = signal_msd[signal_cut_mask]
    signal_hist = flat_histogramdd(signal_msd_cut.numpy(), bins=mybinning)
    signal_hist = normalize_prob(signal_hist)
    signal_hds.append(hellinger_distance(signal_hist, obs_signal_hist))

for background_msd in background_msds:
    background_cut_mask = (background_msd[:, 0] >= -CUT) & (background_msd[:, 0] <= CUT) & (background_msd[:, 1] >= -CUT) & (background_msd[:, 1] <= CUT)
    background_msd_cut = background_msd[background_cut_mask]
    background_hist = flat_histogramdd(background_msd_cut.numpy(), bins=mybinning)
    background_hist = normalize_prob(background_hist)
    background_hds.append(hellinger_distance(background_hist, obs_background_hist))
```
